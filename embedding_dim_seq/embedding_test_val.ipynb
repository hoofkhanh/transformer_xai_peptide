###  Lý do chọn 'post':
#### - Giữ thông tin quan trọng ở đầu chuỗi, vì đầu peptide có thể chứa thông tin quan trọng về tính kháng khuẩn.

```python
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences

padding_type = 'post' # Đệm thêm số 0
trunc_type = 'post'  # Cắt nếu chuỗi dài
```

```python
tokenizer = Tokenizer(char_level=True)
tokenizer.fit_on_texts(data['Sequence'])
vocab_size = len(tokenizer.word_index) + 1
print("Vocab Size: ",vocab_size)
```

```python
tokenizer.word_index
```

```python
# train_sequences  = tokenizer.texts_to_sequences(X_train['Sequence'])
# train_sequence_array = pad_sequences(train_sequences , maxlen= max_length, padding= padding_type, truncating=trunc_type)

# val_sequences = tokenizer.texts_to_sequences(X_val['Sequence'])
# val_sequence_array = pad_sequences(val_sequences , maxlen= max_length, padding= padding_type, truncating=trunc_type)

test_sequences = tokenizer.texts_to_sequences(x_test['Sequence'])
test_sequence_array = pad_sequences(test_sequences , maxlen= max_length, padding= padding_type, truncating=trunc_type)
```

### Embed_dim chuỗi về 50

```python
import os
import numpy as np
import csv
from tensorflow.keras.layers import Embedding, GlobalAveragePooling1D
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input

max_length = 1293
embed_dim = 50

# === Mô hình Embedding + Global Average Pooling ===
input_layer = Input(shape=(max_length,))
embedding_layer = Embedding(input_dim=vocab_size, output_dim=embed_dim, input_length=max_length)(input_layer)
pooled_output = GlobalAveragePooling1D()(embedding_layer)  # [batch_size, 50]

embedding_model = Model(inputs=input_layer, outputs=pooled_output)

# === Tên file CSV ===
csv_file = 'embedded_vector_test_val_sequence_array.csv'
header = ['stt'] + [f'dim_{i}' for i in range(embed_dim)]

# === Kiểm tra số dòng đã xử lý ===
processed_lines = 0
if os.path.exists(csv_file):
    with open(csv_file, 'r') as f:
        processed_lines = sum(1 for _ in f) - 1  # Trừ dòng header

print(f"Đã có {processed_lines} chuỗi được xử lý trước đó.")

# === Bắt đầu xử lý và ghi vào file ===
with open(csv_file, mode='a', newline='') as file:
    writer = csv.writer(file)

    if processed_lines == 0:
        writer.writerow(header)

    for i in range(processed_lines, len(test_sequence_array)):
        seq_nums = np.array(test_sequence_array[i]).reshape(1, -1)
        embedded = embedding_model.predict(seq_nums, verbose=0).squeeze()  # shape: (50,)
        writer.writerow([i + 1] + embedded.tolist())

        if (i + 1) % 100 == 0:
            print(f"Đã xử lý {i + 1}/{len(test_sequence_array)} chuỗi")
```
